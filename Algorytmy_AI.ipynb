```python
import tensorflow as tf

# Tworzenie modelu
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])  # Jedna jednostka i jeden wejściowy cecha
])

# Kompilacja modelu
model.compile(optimizer='sgd', loss='mean_squared_error')

# Dane treningowe
x_train = [1, 2, 3, 4, 5]
y_train = [2, 4, 6, 8, 10]

# Trening modelu
model.fit(x_train, y_train, epochs=100)

# Testowanie modelu
x_test = [6, 7, 8, 9, 10]
y_test = model.predict(x_test)

print(y_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dane treningowe
x_train = np.array([1, 2, 3, 4, 5])
y_train = np.array([2, 4, 6, 8, 10])

# Inicjalizacja wagi i biasu
weight = 0.5
bias = 0.5

# Funkcja aktywacji - funkcja liniowa
def activation(x):
    return weight * x + bias

# Wykres danych treningowych
plt.scatter(x_train, y_train, color='blue', label='Dane treningowe')

# Wykres funkcji aktywacji
x = np.linspace(0, 6, 100)
y = activation(x)
plt.plot(x, y, color='red', label='Sztuczny neuron')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Wizualizacja prostego sztucznego neuronu')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from IPython.display import SVG
from tensorflow.keras.utils import plot_model, model_to_dot

# Tworzenie modelu głębokiej sieci neuronowej
model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(784,)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

# Kompilacja modelu
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Wyświetlanie wizualizacji modelu
plot_model(model, to_file='model.png', show_shapes=True, show_layer_names=True)
SVG(model_to_dot(model).create(prog='dot', format='svg'))
```
