```
# praca z wybranymi bibliotekami Pythona poświęconymi wizualizacji danych

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
x = np.linspace(0, 20, 100)  # dane od 0 do 20 - 100 wartości liczbowych
y = np.random.randn(100, 4)   # liczby wylosowane losowo - 100 liczb i cztery próbki z rozkładu normalnego. W przypadku x i y musi być po 100 (taka sama ilość)
y = np.cumsum(y, axis = 0)     # skumulowana suma, axis - względem osi 0
plt.plot(x, y)                 # prosty wykres w matplotlib
```

```
sns.set()       # połączenie seaborn i matplotlib i wykres jest bardziej szczegółowy
plt.plot(x,y)
```

```
import numpy as np                                     # rozkład rachunków w zależności od dnia
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
df.head()
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'sex')
```

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'box')           # wykres pudełkowy, więcej danych, mediana, kwartyl, etc. Można też: violin (wykres violinowy)
```

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'point') # wykres pokazujący estymację i przedział ufności
```

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'bar')
```

```
sns.catplot(data = df, x = 'day', kind = 'count')
```

```
sns.catplot(data = df, x = 'day', y = 'total_bill', hue = 'sex')      # podzial na płeć
```

```
sns.catplot(data=df.guery('day==['Fri', 'Sat' 'Sun']'), x = 'day', y = 'total_bill', hue = 'sex') # można też użyć dane: 'titanic' !!!
```

```
import numpy as np
import pandas as pd
df = np.random.multivariate_normal(mean=[0, 1], cov=[(1, -0.4), (-0.4, 1)], size = 200)         # rozklad losowy i kowariancja
df = pd.DataFrame(df, columns=["x", "y"])
df.head()
df.describe()
```

```
import seaborn as sns
df.describe()
sns.jointplot(data = df, x = 'x', y = 'y') # połączenie wykresu punktowego z histogramem
```

```
import seaborn as sns
df.describe()
sns.jointplot(data = df, x = 'x', y = 'y', kind = "kde")
```
